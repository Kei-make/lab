```python
import pandas as pd
import glob

csv_files = glob.glob("positioncsv/*.csv")
df = pd.read_csv(csv_files[0],names=("0_x","0_y"))

for i in range(1,9):
    df_temp = pd.read_csv(csv_files[i],names=(f"{i}_x",f"{i}_y"),delim_whitespace=True)
    df = pd.concat([df,df_temp],axis=1)
df
```

```python
df_re = pd.DataFrame([df["0_x"],df["0_y"]]).T

for i in range(1,9):
    df_tempx = df[f"{i}_x"]-df["0_x"]
    df_tempx = df_tempx.rename(index=(f"{i}_x"))
    df_re = pd.concat([df_re,df_tempx],axis=1)
    df_tempy = df[f"{i}_y"]-df["0_y"]
    df_tempy = df_tempy.rename(index=(f"{i}_y"))
    df_re = pd.concat([df_re,df_tempy],axis=1)

df_re
```

```python
df_re1 = df_re
for i in range(14):
    if i%2 == 0:
        df_re1 = df_re1.drop(i)
df_re1 = df_re1[::-1] 
df_tempotary = df_re1.set_axis([7,6,5,4,3,2,1],axis="index")
df_tempotary
```

```python
df_tempotary
```

```python
df_re2 = df_tempotary
df_re2["0_y"] = df_re2["0_y"] -df_re2.iat[-1,1]
df_re2["0_x"] = df_re2["0_x"] -df_re2.iat[-1,0]
df_re2
```

```python
df_tempotary
```

```python
df_xy = df_re2
for i in range(1,9):
    df_xy[f"{i}_y"] = df_xy[f"{i}_y"] + df_xy["0_y"]
    df_xy[f"{i}_x"] = df_xy[f"{i}_x"] + df_xy["0_x"]

df_xy
```

```python
import matplotlib.pyplot as plt
def multiLegend(df, x, y):
    plt.rcParams['font.family'] = "Times new Roman"
    plt.rcParams['xtick.direction'] = 'in'
    plt.rcParams['ytick.direction'] = 'in'
    c_list = ["k", "r", "b", "g", "c", "m", "y","#f781bf","#a65628"]
    l_list = ['-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted']
    y_label = ["","77gal","170gal","274gal","443gal","601gal","705gal","741gal","885gal"]
    fig, ax = plt.subplots(figsize=(5,5))
    plt.subplots_adjust(left=0.17, bottom=0.17, top=0.95, right=0.95)
    plt.xlim(0,6)
    plt.ylim(0,50)
    # x,y
    ax.plot(df["0_x"],df["0_y"],linestyle=l_list[0], color=c_list[0],lw =3,label = y_label[0])
    for i in range(1,len(y)):
        ax.plot(df[x[i]], df[y[i]], linestyle=l_list[9], color=c_list[i],label =y_label[i] )
    plt.legend(loc='lower right', borderaxespad=0.5,
               fontsize=12,prop = {"family":"Meiryo"})
    plt.xlabel("",{"family":"Meiryo","fontsize":"20"})
    plt.ylabel("",{"family":"Meiryo","fontsize":"20"})
    plt.tick_params(labelsize=14)
    plt.show()
    return


multiLegend(df_xy,["0_x","1_x","2_x","3_x","4_x","5_x","6_x","7_x","8_x"], ["0_y","1_y","2_y","3_y","4_y","5_y","6_y","7_y","8_y"])
```

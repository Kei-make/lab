```python
import pandas as pd
import matplotlib.pyplot as plt
df_xy = pd.read_excel("accerate.xlsx",sheet_name = 1,header  = 0)
#コメント

def multiLegend(df, x, y):
    plt.rcParams['font.family'] = "Times new Roman"
    plt.rcParams['xtick.direction'] = 'in'
    plt.rcParams['ytick.direction'] = 'in'
    c_list = ["k", "r", "b", "g", "c", "m", "y","#f781bf","#a65628"]
    l_list = ['-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted']
    y_label = ["振動台加速度","石垣応答加速度","170gal","274gal","443gal","601gal","705gal","741gal","885gal"]
    fig, ax = plt.subplots(figsize=(5,5))
    plt.subplots_adjust(left=0.17, bottom=0.17, top=0.95, right=0.95)
    plt.xlim(114.5,117.5)
    plt.ylim(-500,500)
    # x,yの範囲指定
    for i in range(0,len(y)):
        ax.plot(df[x], df[y[i]], linestyle=l_list[i], color=c_list[i],label =y_label[i])
    plt.legend(loc='lower right', borderaxespad=0.5,
               fontsize=12,prop = {"family":"Meiryo"})
    plt.xlabel("測定時間",{"family":"Meiryo","fontsize":"20"})
    plt.ylabel("加速度(gal)",{"family":"Meiryo","fontsize":"20"})
    plt.tick_params(labelsize=14)
    plt.show()
    return

y_label = [{"初期位置":"0"},{"170gal":"2"},{"274gal":"3"},{"601gal":"5"},{"705gal":"6"},{"885gal":"8"}]
xs ="time"
ys = ["ac1","rockac1"]
multiLegend(df_xy,xs,ys)
```
